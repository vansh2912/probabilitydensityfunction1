```python
import numpy as np
import pandas as pd

data = pd.read_csv("data.csv", encoding='latin1', low_memory=False)

# Extract NO2 features only
no2_column = [col for col in data.columns if col.lower() == "no2"][0]
x = data[no2_column].dropna().values


# Step 1: Transformation
# z = x + a_r sin(b_r x)

r = 102303641

a_r = 0.05 * (r % 7)
b_r = 0.3 * ((r % 5) + 1)

z = x + a_r * np.sin(b_r * x)


# Step 2: Learn PDF Parameters
# p(z) = c * exp(-λ (z - μ)^2)

# Estimate μ and variance from transformed data
mu = np.mean(z)
sigma2 = np.var(z)

# Compute λ and c
lam = 1 / (2 * sigma2)
c = np.sqrt(lam / np.pi)

print("Roll Based Parameters")
print("a_r =", a_r)
print("b_r =", b_r)

print("Learned PDF Parameters")
print("Mean (μ) =", mu)
print("Variance (σ^2) =", sigma2)
print("Lambda (λ) =", lam)
print("c =", c)

# Predicted Probability Function


def predicted_pdf(z_value):
    return c * np.exp(-lam * (z_value - mu) ** 2)

# Example: compute probability for first transformed value
print("\nSample Probability Value:", predicted_pdf(z[0]))
```
